# Model
$$
\begin{align}
p(\boldsymbol{Y},\boldsymbol{Z},\boldsymbol{S},\boldsymbol{\Lambda},\boldsymbol{R},\boldsymbol{\pi})=&p(\boldsymbol{Y}|\boldsymbol{Z},\boldsymbol{S},\boldsymbol{\Lambda},\boldsymbol{R})p(\boldsymbol{Z})p(\boldsymbol{S}|\boldsymbol{\pi})p(\boldsymbol{\Lambda})p(\boldsymbol{R})p(\boldsymbol{\pi})
\\
p(\boldsymbol{Y}|\boldsymbol{Z},\boldsymbol{S},\boldsymbol{\Lambda},\boldsymbol{R})&=\prod_{t=1}^T\prod_{d=1}^D\prod_{k=1}^K\mathcal{N}(\boldsymbol{y}_{d,t}|\boldsymbol{\Lambda}_d^{(k)^T}\boldsymbol{z}_t,\boldsymbol{R}_d^{(k)^{−1}})^{s_t^{(k)}}
\\
p(\boldsymbol{S}|\boldsymbol{\pi})&=\prod_{t=1}^{T}\mathrm{Cat}(s_t|\boldsymbol{\pi})
\\
p(\boldsymbol{\Lambda})&=\prod_{d=1}^{D}\prod_{k=1}^{K}\mathcal{N}(\boldsymbol{\Lambda}_d^{(k)}|\boldsymbol{\mu}^{(k)}_{\boldsymbol{\Lambda}_d},\boldsymbol{\Sigma}^{(k)}_{\boldsymbol{\Lambda}_d})
\\
p(\boldsymbol{R})&=\prod_{d=1}^{D}\prod_{k=1}^{K}\mathrm{Gam}(\boldsymbol{R}_{d}^{(k)}|a_{R_d}^{(k)},b_{R_d}^{(k)})
\\
p(\boldsymbol{Z})&=\prod_t^T \mathcal{N}(z_{t}|\mu_{z},\Sigma_{z})=\prod_t^T \prod_{l=1}^{L + 1} \mathcal{N} (z_{l,t}|\mu_{z_l},\sigma_{z_l}^2)
\\
&\mu_{z_l}=\left\{\begin{matrix}
0,&l&=& 1,\cdots,L
\\
1,&l&=&L + 1
\end{matrix}\right.
\\
&\sigma_{z_l}^{2}=\left\{\begin{matrix}
1,&l&=&1,\cdots,L
\\
\epsilon,&l&=&L+1 
\end{matrix}\right.
\end{align}
$$
(R is diagonal)



# Variational Inference

$$
\begin{align}
p(\boldsymbol{Z},\boldsymbol{S},\boldsymbol{\Lambda},\boldsymbol{R},\boldsymbol{\pi}) \approx & q(\boldsymbol{Z},\boldsymbol{S})q(\boldsymbol{\Lambda})q(\boldsymbol{R})q(\boldsymbol{\pi})
\\
\ln p(\boldsymbol{Z},\boldsymbol{S}) &= \langle \ln p(\boldsymbol{Y}|\boldsymbol{Z},\boldsymbol{S},\boldsymbol{\Lambda},\boldsymbol{R})\rangle_{q(\boldsymbol{\Lambda},\boldsymbol{R})}+ \ln p(\boldsymbol{Z})+\langle p(\boldsymbol{S}|\boldsymbol{\pi})\rangle_{q(\boldsymbol{\pi})} + C
\\
\ln q(\boldsymbol{\Lambda})&=\langle \ln p(\boldsymbol{Y}|\boldsymbol{Z},\boldsymbol{S},\boldsymbol{\Lambda},\boldsymbol{R})\rangle_{q(\boldsymbol{Z},\boldsymbol{S}),q(\boldsymbol{R})}+\ln p(\boldsymbol{\Lambda})+C
\\
\ln q(\boldsymbol{R})&=\langle \ln p(\boldsymbol{Y}|\boldsymbol{Z},\boldsymbol{S},\boldsymbol{\Lambda},\boldsymbol{R})\rangle_{q(\boldsymbol{Z},\boldsymbol{S}),q(\boldsymbol{\Lambda})}+\ln q(\boldsymbol{R})+C
\\
\ln q(\boldsymbol{\pi})&=\langle\ln p(\boldsymbol{S}|\boldsymbol{\pi})\rangle_{q(\boldsymbol{S})}+\ln p(\boldsymbol{\pi}) +C
\end{align}
$$


# q(Z,S)
$$
\begin{align}
\ln p(\boldsymbol{Z},\boldsymbol{S}) =& \langle \ln p(\boldsymbol{Y}|\boldsymbol{Z},\boldsymbol{S},\boldsymbol{\Lambda},\boldsymbol{R})\rangle_{q(\boldsymbol{\Lambda},\boldsymbol{R})}+ \ln p(\boldsymbol{Z})+\langle p(\boldsymbol{S}|\boldsymbol{\pi})\rangle_{q(\boldsymbol{\pi})} + C
\\
=&\sum_{t=1}^{T}\sum_{k=1}^{K}\sum_{d=1}^{D}−\frac{1}{2}s_t^{(k)}\langle \{(y_{d,t} − \boldsymbol{\Lambda}_d^{(k)^T} \boldsymbol{z}_t^{(k)})^2 R^{(k)}_d-\ln R^{(k)}_d+D\ln 2\pi\}\rangle 
\\
&+\sum_{t=1}^{T}\sum_{k=1}^{K}(\boldsymbol{z}^{(k)}_{t} - \boldsymbol{\mu}_{z})^T \boldsymbol{\Sigma}_{z}^{-1} (\boldsymbol{z}^{(k)}_{t} - \boldsymbol{\mu}_{z})+\ln |\boldsymbol{\Sigma}_z| + L \ln 2\pi 
\\
&+ \sum_{t=1}^{T}\sum_{k=1}^{K} s_t^{(k)}\langle\ln\pi^{(k)}\rangle +C
\\
=&\sum_{t=1}^{T}\sum_{k=1}^{K}\sum_{d=1}^{D}−\frac{1}{2}s_t^{(k)} \{(y_{d,t}^2 \langle R^{(k)}_d \rangle − 2 y_{d,t}\langle\boldsymbol{\Lambda}_d^{(k)^T}\rangle \boldsymbol{z}_t^{(k)}\langle R^{(k)}_d\rangle + (\langle \boldsymbol{\Lambda}_d^{(k)^T}\rangle \boldsymbol{z}_t^{(k)})(\langle \boldsymbol{ \Lambda}_d^{(k)^T}\rangle  \boldsymbol{z}_t^{(k)})\langle R^{(k)}_d \rangle-\langle \ln R^{(k)}_d\rangle +D\ln 2\pi\}
\\
&+\sum_{t=1}^{T}\sum_{k=1}^{K}\boldsymbol{z}^{(k)^T}_{t} \boldsymbol{\Sigma}_{z}^{-1}\boldsymbol{z}^{(k)}_{t} - 2 \boldsymbol{z}^{(k)^T}_{t} \boldsymbol{\Sigma}_{z}^{-1}\boldsymbol{\mu}_{z} + \boldsymbol{\mu}_{z}^T \boldsymbol{\Sigma}_{z}^{-1}\boldsymbol{\mu}_{z}+\ln |\boldsymbol{\Sigma}_z| + L \ln 2\pi 
\\
&+ \sum_{t=1}^{T}\sum_{k=1}^{K} s_t^{(k)}\langle\ln\pi^{(k)}\rangle +C
\\
=& \ln q(\boldsymbol{Z}|\boldsymbol{S}) + \ln q(\boldsymbol{S})
\end{align}
$$
## q(Z|S)
$$
\begin{align}
\ln q(\boldsymbol{Z}|\boldsymbol{S})=& \sum_{t=1}^{T}\sum_{k=1}^{K} s_t^{(k)} \mathcal{N}(\boldsymbol{z}_t^{(k)}|\hat{\boldsymbol{\mu}}_z^{(k)}, \hat{\boldsymbol{\Sigma}}_z^{(k)}) + C
\\
&\hat{\boldsymbol{\Sigma}}_z^{(k)^{-1}}=\boldsymbol{\Sigma}_z^{-1} + \Sigma_{d=1}^{D} \langle \boldsymbol{\Lambda}_d^{(k)}\boldsymbol{\Lambda}_d^{(k)^T}\rangle\langle R_d^{(k)}\rangle
\\
&\hat{\boldsymbol{\mu}}_z^{(k)}=\hat{\boldsymbol{\Sigma}}_z^{(k)}[\Sigma_{d=1}^{D} y_{d,t}\langle R_d^{(k)}\rangle\langle\boldsymbol{\Lambda}_d^{(k)}\rangle +\boldsymbol{\Sigma}_z^{-1}\boldsymbol{\mu}_z]
\end{align}
$$
## q(S)
$$
\begin{align}
\ln q(\boldsymbol{S})=& \sum_{t=1}^{T}\mathrm{Cat}(\boldsymbol{s}_t|\hat{\boldsymbol{\alpha}}_{\pi}) + C
\\
\hat{\eta}&\propto \exp[-\frac{1}{2}\{\sum_{d=1}^D y_{d,t}^2 \langle R_d^{(k)}\rangle - \sum_{d=1}^D \langle \ln R_d^{(k)}\rangle + \boldsymbol{\mu_{z}}^T\boldsymbol{\Sigma}_{z}^{-1}\boldsymbol{\mu}_{z}+\hat{\boldsymbol{\mu}}_{z_t}^{(k)^T}\hat{\boldsymbol{\Sigma}}_{z}^{-1}\hat{\boldsymbol{\mu}}_{z_t}^{(k)}+\ln|\boldsymbol{\Sigma}_{z}|+\ln|\hat{\boldsymbol{\Sigma}}_{z}|+D\ln 2\pi+\}\langle \ln \pi_k\rangle]
\end{align}
$$

# q(Λ)
$$
\begin{align}
&\ln q(\boldsymbol{\Lambda})=\langle \ln p(\boldsymbol{Y}|\boldsymbol{Z},\boldsymbol{S},\boldsymbol{\Lambda},\boldsymbol{R})\rangle_{q(\boldsymbol{Z},\boldsymbol{S}),q(\boldsymbol{R})}+\ln p(\boldsymbol{\Lambda})+C
\\
&=\sum_{t=1}^{T}\sum_{k=1}^{K}\sum_{d=1}^{D}−\frac{1}{2}\langle s_t^{k}\{(y_{td} − \boldsymbol{\Lambda}_d^{(k)^T} \boldsymbol{z}_t^{(k)})^2 R^{(k)}_d\}\rangle + (\boldsymbol{\Lambda}^{(k)}_d - \boldsymbol{\mu}_{\Lambda_d}^{(k)})^T \boldsymbol{\Sigma}_{d}^{(k)^{-1}} (\boldsymbol{\Lambda}^{(k)}_d - \boldsymbol{\mu}_{\Lambda_d}^{(k)})+C
\\
&= \sum_{t=1}^{T}\sum_{k=1}^{K}\sum_{d=1}^{D}−\frac{1}{2} \{\boldsymbol{\Lambda}^{(k)^T}_d\langle R_d^{(k)}\rangle\langle s_t^{(k)}\boldsymbol{z}_d^{(k)}\boldsymbol{z}_d^{(k)^T}\rangle \boldsymbol{\Lambda}_d^{(k)}-2\boldsymbol{\Lambda}^{(k)^T}_d\langle R_d^{(k)}\rangle \langle s_t^{(k)}\boldsymbol{z}_d^{(k)^T}\rangle + \{\boldsymbol{\Lambda}^{(k)^T}_d \boldsymbol{\Sigma}_{\Lambda_d}^{(k)^{-1}} \boldsymbol{\Lambda}_d^{(k)} -2\boldsymbol{\Lambda}_d^{(k)^{T}} \boldsymbol{\Sigma}_{\Lambda_d}^{(k)^{-1}}\boldsymbol{\mu}_{\Lambda_d}^{(k)}\} + C
\\
&= \sum_{t=1}^{T}\sum_{k=1}^{K}\sum_{d=1}^{D}−\frac{1}{2} \{\boldsymbol{\Lambda}^{(k)^T}_d (\langle R_d^{(k)}\rangle \langle s_t^{(k)}\boldsymbol{z}_d^{(k)}\boldsymbol{z}_d^{(k)^T}\rangle + \boldsymbol{\Sigma}_{\Lambda_d}^{(k)^{-1}})\boldsymbol{\Lambda}_d^{(k)}-2\boldsymbol{\Lambda}^{(k)^T}_d (\langle R_d^{(k)}\rangle \langle s_t^{(k)}\boldsymbol{z}_d^{(k)}\rangle
 + \boldsymbol{\Sigma}_{\Lambda_d}^{(k)^{-1}}\boldsymbol{\mu}_{\Lambda_d}^{(k)})\} + C
 \\
 &=\sum_{k=1}^{K}\sum_{d=1}^{D} \ln \mathcal{N}(\boldsymbol{\Lambda}^{(k)^T}_d|\hat{\boldsymbol{\mu}}_{\Lambda_d}^{(k)},\hat{\boldsymbol{\Sigma}}_{\Lambda_d}^{(k)^{-1}})
\end{align}
$$
$$
\left\{\begin{align}
\hat{\boldsymbol{\Sigma}}_{\Lambda_d}^{(k)^{-1}}&= \langle R_d^{(k)}\rangle \sum_{t=1}^{T}\langle s_t^{(k)}\boldsymbol{z}_d^{(k)}\boldsymbol{z}_d^{(k)^T}\rangle + \boldsymbol{\Sigma}_{\Lambda_d}^{(k)^{-1}}
\\
\hat{\boldsymbol{\mu}}_{\Lambda_d}^{(k)}&=\hat{\boldsymbol{\Sigma}}_{\Lambda_d}^{(k)}[\langle R_d^{(k)}\rangle \sum_{t=1}^{T}\langle s_t^{(k)}\boldsymbol{z}_d^{(k)}\rangle
 + \boldsymbol{\Sigma}_{\Lambda_d}^{(k)^{-1}}\boldsymbol{\mu}_{\Lambda_d}^{(k)}]
\end{align}\right.
$$

# q(R)
$$
\begin{align}
&\ln q(\boldsymbol{R})=\langle \ln p(\boldsymbol{Y}|\boldsymbol{Z},\boldsymbol{S},\boldsymbol{\Lambda},\boldsymbol{R})\rangle_{q(\boldsymbol{Z},\boldsymbol{S}),q(\boldsymbol{\Lambda})}+\ln q(\boldsymbol{R})+C
\\
&=\sum_{t=1}^{T}\sum_{k=1}^{K}\sum_{d=1}^{D}−\frac{1}{2}\langle s_t^{(k)}\{(y_{td} − \boldsymbol{\Lambda}_d^{(k)^T} \boldsymbol{z}_t^{(k)})^2 R^{(k)}_d-\ln R^{(k)}_d\}\rangle + \sum_{k=1}^{K}\sum_{d=1}^{D} (a_{R_d}^{(k)}-1)\ln R_d^{(k)}-b_{R_d}^{(k)}R_d^{(k)}+C
\\
&=\sum_{k=1}^{K}\sum_{d=1}^{D} [(a_{R_d}^{(k)}+\frac{1}{2}\sum_{t=1}^{T}\langle s_t^{(k)}\rangle-1)\ln R_d^{(k)}-\{b_{R_d}^{(k)}+\frac{1}{2}\sum_{t=1}^{T}\langle s_t^{(k)}(y_{td} − \boldsymbol{\Lambda}_d^{(k)^T} \boldsymbol{z}_t^{(k)})^2 \rangle\}R_d^{(k)}]+C
\\
&=\sum_{k=1}^{K}\sum_{d=1}^{D} [(a_{R_d}^{(k)}+\frac{1}{2}\sum_{t=1}^{T}\langle s_t^{(k)}\rangle-1)\ln R_d^{(k)}-\{b_{R_d}^{(k)}+\frac{1}{2}\sum_{t=1}^{T}\{\langle s_t^{(k)}\rangle y_{td}^2 - 2y_{td} \langle \boldsymbol{\Lambda}_d^{(k)^T}\rangle \langle s_t^{(k)}\boldsymbol{z}_t^{(k)}\rangle +\langle s_t^{(k)}(\boldsymbol{\Lambda}_d^{(k)^T} \boldsymbol{z}_t^{(k)})^2 \rangle\}\}R_d^{(k)}]+C
\\
&=\sum_{k=1}^{K}\sum_{d=1}^{D} (a_{R_d}^{(k)}+\frac{1}{2}\sum_{t=1}^{T}\langle s_t^{(k)}\rangle-1)\ln R_d^{(k)}-[b_{R_d}^{(k)}+\frac{1}{2}\sum_{t=1}^{T}\{\langle s_t^{(k)}\rangle y_{td}^2 - 2y_{td} \langle \boldsymbol{\Lambda}_d^{(k)^T}\rangle \langle s_t^{(k)}\boldsymbol{z}_t^{(k)}\rangle +\mathrm{Tr}(\langle s_t^{(k)}\boldsymbol{z}_t^{(k)}\boldsymbol{z}_t^{(k)^T}\rangle \langle \boldsymbol{\Lambda}_d^{(k)}\boldsymbol{\Lambda}_d^{(k)^T} \rangle\} ] R_d^{(k)}+C
\\
&=\sum_{k=1}^{K}\sum_{d=1}^{D} \mathrm{Gam}(R_d^{(k)}\mid \hat{a}_{R_d}^{(k)},\hat{b}_{R_d}^{(k)})+C
\end{align}
$$

$$
\left\{\begin{align}
\hat{a}_{R_d}^{(k)} &= a_{R_d}^{(k)}+\frac{1}{2}\sum_{t=1}^{T}\langle s_t^{(k)}\rangle
\\
\hat{b}_{R_d}^{(k)} &= b_{R_d}^{(k)}+\frac{1}{2}\sum_{t=1}^{T}\{\langle s_t^{(k)}\rangle y_{d,t}^2 - 2y_{td} \langle \boldsymbol{\Lambda}_d^{(k)^T}\rangle \langle s_t^{(k)}\boldsymbol{z}_t^{(k)}\rangle +\mathrm{Tr}(\langle s_t^{(k)}\boldsymbol{z}_t^{(k)}\boldsymbol{z}_t^{(k)^T}\rangle \langle \boldsymbol{\Lambda}_d^{(k)}\boldsymbol{\Lambda}_d^{(k)^T} \rangle\}
\end{align} \right.
$$

# q(π)
$$
\begin{align}
\ln q(\boldsymbol{\pi})&=\langle\ln p(\boldsymbol{S}|\boldsymbol{\pi})\rangle_{q(\boldsymbol{S})}+\ln p(\boldsymbol{\pi}) +C
\\
&=\sum_{t=1}^{T} \sum_{k=1}^{K} \langle s_t^{(k)}\rangle \ln\pi^{(k)} + \sum_{k=1}^{K}(\alpha_{\pi}^{(k)} − 1)\ln\pi^{(k)} + C
\\
&=\sum_k^K(\sum_t^T\langle s_t^{(k)}\rangle +\alpha_{\pi}^{(k)}−1)\ln\pi^{(k)}+C
\\
&=\ln\mathrm{Dir}(\pi|\hat{\alpha}_{\pi})
\end{align}
$$
